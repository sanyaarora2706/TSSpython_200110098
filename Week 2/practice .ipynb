```python
import matplotlib.pyplot as plt
```

```python
x=[4,9,6]
y=[6,7,8]
plt.plot(x,y)
plt.show
```

```python
import numpy as np
xpoints = np.array([1, 5])
ypoints = np.array([4, 3])

plt.plot(xpoints, ypoints, 'o')
plt.show()
# To plot the above with red circles
plt.plot(xpoints, ypoints, 'yo')
plt.show()
```

```python
x=[5,9,111]
y=[7,9,3]
plt.plot(x,y,'o:b')
plt.show
```

```python
x=[4,7,9,5]
y=[2,7,9,9]
plt.bar(x,y)
plt.show
```

```python
x=np.arange(0,3*np.pi,0.1)
ysin=np.sin(x)
ycos=np.cos(x)
plt.plot(x,ysin,ls='dotted')
plt.plot(x,ycos,'o')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('graph of sin and cos')
plt.legend(['sin','cos'])
plt.show()
```

```python
x=[4,5,6]
y=[6,7,7]
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.normal(100,20,200)
plt.hist(x)
plt.show()
```

```python
y=np.array([5,49,77])
mylabels=["apples","oranges","bananas"]
plt.pie(y,labels=mylabels)
plt.show()
```
